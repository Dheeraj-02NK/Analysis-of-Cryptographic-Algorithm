```python
from random import randint
import time
import random
import string
str_list = []

Alpahbets = ' A!1B@2C#D3$E%4F^G5&H*6I(J7)K_8L-M9+N=0O:P;Q\'R"S?T/U>V<W.X,Y`Z~'

def encrypt(text,key):
    text = text.upper()
    cipher_text = ''
    for index, char in enumerate(text):
        key_index = key[index]
        char_index = Alpahbets.find(char)
        cipher_text += Alpahbets[(char_index + key_index) % len(Alpahbets)]
    return  cipher_text

def random_sequence(text):
    random = []
    for _ in range(len(text)):
        random.append(randint(0,len(Alpahbets)))
    return random

if __name__ == '__main__':
    
    for length in range(5, 2501, 5):
        rand_string = ''.join(random.choices(string.ascii_uppercase, k=length))
        str_list.append(rand_string)
        
        start_time = time.time()
        for message in str_list:
            seq = random_sequence(message)
            cipher_text = encrypt(message, seq)
        end_time = time.time()
        print("{:.2f}".format((end_time - start_time)*1000))
    
```
