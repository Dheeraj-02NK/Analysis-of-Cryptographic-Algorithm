```python
import random
from math import floor
from math import sqrt
import time
import random
import string
str_list = []

RANDOM_START = 1e3
RANDOM_END = 1e5


def is_prime(num):
    if num < 2:
        return False
    
    if num == 2:
        return True

    if num % 2 == 0:
        return False

    for i in range(3, floor(sqrt(num))):
        if num % i == 0:
            return False
    return True

def gcd(a, b):

    while b != 0:
        a, b = b, a % b
    return a

def modular_inverse(a, b):
    if a == 0:
        return b, 0, 1
    div, x1, y1 = modular_inverse(b % a, a)
    x = y1 - (b // a) * x1
    y = x1
    return div, x, y


def generate_large_prime(start=RANDOM_START, end=RANDOM_END):
    num = random.randint(start, end)
    while not is_prime(num):
        num = random.randint(start, end)
    return num


def generate_rsa_keys():
    p = generate_large_prime()
    q = generate_large_prime()
    n = p * q
    phi = (p-1)*(q-1)
    e = random.randrange(1, phi)
    while gcd(e, phi) != 1:
        e = random.randrange(1, phi)
    d = modular_inverse(e, phi)[1]
    return (d, n), (e, n)



def encrypt(public_key, plain_text):
    e, n = public_key
    cipher_text = []
    for char in plain_text:
        a = ord(char)
        cipher_text.append(pow(a, e, n))
    return cipher_text

if __name__ == '__main__':
    
    for length in range(5, 2501, 5):
        rand_string = ''.join(random.choices(string.ascii_uppercase, k=length))
        str_list.append(rand_string)

        start_time = time.time()
        for message in str_list:
            private_key, public_key = generate_rsa_keys()
            cipher = encrypt(public_key, message)
            
        end_time = time.time()
        print("{:.2f}".format((end_time - start_time)*1000))
    
```
